```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import math
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data_df = pd.merge(mouse_metadata, study_results, how="outer", on="Mouse ID")
# Display the data table for preview
combined_data_df
```

```python
# Checking the number of mice.
mouse = combined_data_df["Mouse ID"].value_counts()
number_of_mice = len(mouse)
number_of_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
# Optional: Get all the data for the duplicate mouse ID. 
dupemice = combined_data_df[combined_data_df.duplicated(subset=["Mouse ID", "Timepoint"], keep=False)]["Mouse ID"].unique()
for i in dupemice:
    print(i)
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
combined_data_df = combined_data_df[~combined_data_df.rename(columns={"Mouse ID":"MouseID"}).MouseID.isin(dupemice)]
combined_data_df
```

```python
# Checking the number of mice in the clean DataFrame.
len(combined_data_df)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method is the most straighforward, creating multiple series and putting them all together at the end.
summarystat_df = pd.DataFrame(combined_data_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean())
summarystat_df = summarystat_df.rename(columns={"Tumor Volume (mm3)": "Tumor Vol. Mean"})

summarystat_df["Tumor Vol. Median"] = pd.DataFrame(combined_data_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].median())
summarystat_df["Tumor Vol. Variance"] = pd.DataFrame(combined_data_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].var())
summarystat_df["Tumor Vol. Standard Deviation"] = pd.DataFrame(combined_data_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].std())
summarystat_df["Tumor Vol. SEM"] = pd.DataFrame(combined_data_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem())
summarystat_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method produces everything in a single groupby function
summarystat2_df = pd.DataFrame(combined_data_df[["Drug Regimen", "Tumor Volume (mm3)"]])

summarystat2_df = summarystat2_df.groupby("Drug Regimen").agg(["mean", "median", "var", "std", "sem"])
summarystat2_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
drugfilter = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
drugfilter_df = pd.DataFrame(combined_data_df[combined_data_df.rename(columns = {"Drug Regimen":"DR"}).DR.isin(drugfilter)]).copy()

# Start by getting the last (greatest) timepoint for each mouse
maxtimepoint = drugfilter_df.groupby(['Mouse ID', 'Drug Regimen'],as_index=False)['Timepoint'].max()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
lasttimepoint=(maxtimepoint['Mouse ID']+maxtimepoint['Drug Regimen']+maxtimepoint['Timepoint'].map(str))

# adding all into single column
drugfilter_df["FilterRow"] = (drugfilter_df["Mouse ID"]+drugfilter_df['Drug Regimen']+drugfilter_df['Timepoint'].map(str))

# filter the data for 4 drugs
drugfilter_df = drugfilter_df[drugfilter_df.FilterRow.isin(lasttimepoint)].reset_index()
drugfilter_df
```

```python
# Put treatments into a list for for loop (and later for plot labels
# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = drugfilter_df['Tumor Volume (mm3)']

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
quartiles = tumor_vol_data.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq
    
# Locate the rows which contain mice on each drug and get the tumor volumes
# add subset 
nsamples = 25
div = 20
samples = [drugfilter_df.sample(div) for x in range(0,nsamples)]
    
# Determine outliers using upper and lower bounds
lowerbound = lowerq - (1.5*iqr)
upperbound = upperq - (1.5*iqr)

# printing it all
print(f' The lower quartile of the Tumor volume is: {lowerq} mm3')
print(f' The upper quartile of the Tumor volume is: {upperq} mm3')
print(f' The interquartile range of the Tumor volume is: {iqr}')
print(f' The median of the Tumor volume is: {quartiles[0.5]} mm3')
print(f' The values below {lowerbound} could be outliers')
print(f' The values above {upperbound} could be outliers')
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig1,ax1 = plt.subplots()
ax1.set_title("Final tumor volume for the 4 drug regimen of interest")
ax1.set_ylabel('Tumour Volume (mm3)')
ax1.boxplot(tumor_vol_data)
x_axis = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
tick_locations = [value for values in x_axis]
plt.xticks(tick_locations,rotations=45, horizontalalignment='right')
plt.show()

#drugregimen = drugfilter_df.loc[drugfilter_df["Drug Regimen"]]["Tumor Volume(mm3)"]
#plt.figure(figsize=(12,8))
#plt.boxplot(drugregimen,flierprops=outliermarker)
#plt.title('Final Tumor Volume')
#plt.xlabel('Drug Regimen')
#plt.ylabel('Tumor Volume (mm3)')
#plt.xticks(x_axis_element,xticks_labels)
#plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
mousecapo = combined_data_df.loc[combined_data_df['Drug Regimen'] == 'Capomulin']['Mouse ID'].unique()
sampleindex = np.random.randint(0,len(mousecapo)-1)
mouseid = mousecapo[sampleindex]
sampledata = combined_data_df.loc[combined_data_df["Mouse ID"]==mouseid]

# format the yticks
minvol = math.floor(sampledata['Tumor Volume (mm3)'].min())
maxvol = math.ceil(sampledata['Tumor Volume (mm3)'].max())
variance = (maxvol-minvol)/10
yticklabel = np.arange(minvol-variance,maxvol+variance,variance)

#generate line plot for the random capo mouse
plt.figure(figsize=(12,8))
plt.plot(sampledata['Timepoint'], sampledata['Tumor Volume (mm3)'], marker='o', color='blue')

#update labels and title
plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume (mm3)')
plt.title('TIme point vs Tumor Volume - Mouse ID: ' + mouseid)

#updaeticks
plt.ylim(minvol-variance,maxvol+variance)
plt.yticks(yticklabel)

#show the plot
plt.grid()
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
capo = combined_data_df.loc[combined_data_df['Drug Regimen'] == "Capomulin"]

# getting the average tumor
capo = capo.groupby("Weight (g)", as_index=False)['Tumor Volume (mm3)'].mean()
mouseweight = capo.iloc[:,0]
averagetumor = capo.iloc[:,1]

# creating the scatter plot
plt.figure(figsize=(12,8))
caposcatter = plt.scatter(mouseweight, averagetumor, marker='o', facecolor='blue', s=capo['Tumor Volume (mm3)']*3)

# add labels and title
plt.xlabel('Mouse weight (g)')
plt.ylabel("Average Tumor Volume (mm3)")
plt.title('Mouse weight vs Average Tumor Volume (mm3)')

#plot
caposcatter2 = caposcatter
plt.show(caposcatter)
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation = st.pearsonr(mouseweight,averagetumor)
print(f"The correlation between both factors is {round(correlation[0],2)}")

# linear regression model
x_values = mouseweight
y_values = averagetumor
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(f'The linear regression model is {line_eq}')
```
